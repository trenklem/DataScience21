```python
#3.1: import the numpy library
import numpy as np
```

```python
#3.2: create a random 10x10 array named A
A = np.random.rand(10,10)
A
```

```python
#3.3: return only the odd rows of A
A[0:10:2,:] # start:stop:stepsize , all
```

```python
#3.4: perform a matrix multiplication of A with it's transposed
A.dot(A.T)
```

```python
#3.5: compute the matrix inverse of A
A_inv=np.linalg.inv(A)
np.isclose(A.dot(A_inv),np.identity(10))#test for identity AA^-1
```

```python
#same in one line with where
np.where(A>0.5, A/2, A) # np.where(condition, Array to take value if condition is true, Array to take Value if condition is false )
```

```python
# if we need only the values to change:
mask=A>0.5 #generate a mask with indix values
A[mask]/2 #manipulate Array at mask locations

```
